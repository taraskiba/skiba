```python
col = ee.ImageCollection("NASA/ORNL/DAYMET_V4")
col = col.filterDate(svalue, evalue)
img = col.median()
img

coordinates = pd.read_csv("../data/selected_coordinates.csv")
gdf = gpd.GeoDataFrame(
    coordinates,
    geometry=gpd.points_from_xy(coordinates.LON, coordinates.LAT),
    crs="EPSG:4326",  # Directly set CRS during creation
)
geojson = gdf.__geo_interface__
fc = gm.geojson_to_ee(geojson)
out_dir = os.path.join(os.path.expanduser("~"), "Downloads")
output_file = f"test.csv"
out_path = os.path.join(out_dir, output_file)
# Retrieve data from the image using sampleRegions
sampled_data = gm.extract_values_to_points(fc, img, out_path)
```
